# 深度学习硬件：TPU和其他

## DSP 数字信号处理器
本质上是一种专门用于**实时数字信号处理**的微处理器

靠极致的硬件优化，以最低功耗和最快速度完成特定类型的数学运算

CPU（冯·诺依曼）：一条内存通道，既取指令又取数据。（存在瓶颈）

DSP（改良哈佛结构）：程序总线 + 数据总线完全分开。在取下一个指令的同时，可以同时访问两个不同的数据块。能在一个时钟周期内完成“乘法+加法”并同时取好下一次运算的数据，效率极高。

## FPGA 现场可编程门阵列
FPGA 内部是一大片空白的“可编程逻辑块”（LUT查找表、触发器）和可编程连线。

写的代码（Verilog/VHDL，硬件描述语言）会被综合成物理电路图，然后烧录进去，把这些空白积木搭建成你想要的专用电路（硬件即软件）。

## AI ASIC ：AI专用集成电路
为矩阵乘法等AI运算量身定制且出厂即固化，追求极致能效比的专用芯片。

## Systolic Array 脉动阵列
让数据像流水线脉冲一样在相邻计算单元间有节奏流动，从而大幅减少内存访问的特定数据流架构
