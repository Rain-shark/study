# 数据操作和预处理

## 数据操作
深度学习存储和操作数据的主要接⼝是张量（n维数组）。它提供了各种功能，包括基本数学运算、⼴播、索引、切⽚、内存节省和转换其他Python对象。

### 索引和切片
第一个元素的索引是0，最后一个元素索引是-1。
可以用[-1]选择最后一个元素，可以用[1:3]选择第二个和第三个元素
eg.
    访问一个具体元素  [1,2]
    访问一行元素  [1,:]
    访问一列元素  [:,1]
    访问连续子区域   [1:3,1:]
    访问间隔区域：如每隔3行2列访问元素  [::3,::2]

张量表示由⼀个数值组成的数组，这个数组可能有多个维度。具有一个轴的张量对应数学上的向量（vector）； 具有两个轴的张量对应数学上的矩阵（matrix）；具有两个轴以上的张量没有特殊的数学名称。

### 数据运算

+、-、*、/、**都是按照元素运算

这里的*是对应元素相乘，和数学上的矩阵乘法不同

```python
x = torch.tensor([1.0,2,4,8])
y=torch.tensor([2,2,2,2])
x*y,x-y,x+y,x/y,x**y
```

```python
x1 = torch.tensor([[1,2],[4,8]])
y1 =torch.tensor([[2,2],[2,2]])
x1@y1   #矩阵乘法（行*列）
x1*y1   #哈达玛积 （逐元素相乘）
```

```python
torch.exp(x)
```

把多个张量连接在一起，用torch.cat()，dim指定维度（dim=0是按行连接，dim=1是按列连接）

### 广播机制

也可以直接reshape(3，1)

```python
a =torch.arange(3).reshape((3,1))
b = torch.arange(2).reshape((1,2))
a,b
```

```python
a+b
```

### 节省内存
运行一些操作可能导致为新结果分配内存
如果Y=Y+X，将取消引用Y指向的张量，指向新分配的内存处张量

```python
before = id(y)
y=y+x
id(y)==before
```

原地操作：可使用切片表示法将操作的结果分配给先前分配的数组
Y[:]=<expression>

```python
z=torch.zeros_like(y)
print('id(z):',id(z))
z[:]=x+y
print('id(z):',id(z))
```

也可以使用 y+=x 来减少内存开销

```python
before = id(y)
y+=x
id(y)==before
```

## 数据预处理

### 创建.csv文件（逗号分隔符）

```python
import os
os.makedirs(os.path.join('..','data'),exist_ok=True)   
#创建文件夹在父文件夹下，文件存在也不报错
data_file= os.path.join('..','data','house_tiny.csv')
#文件路径
with open(data_file,'w')as f:    
#写入文件，若文件不存在则创建，若存在则覆盖
    f.write('NumRooms,Alley,Price\n')  #写入表头
    f.write('NA,Pave,127500\n')         #写入数据
    f.write('2,NA,106000\n')
    f.write('4,NA,178100\n')
    f.write('NA,NA,140000\n')
```

### 从创建的文件中读取数据

```python
import pandas as pd
data=pd.read_csv(data_file)
print(data)
```

### 处理缺失值

“NaN”
1. 插值法：用替代值弥补缺失值
2. 删除法：忽略缺失值

```python
inputs,outputs= data.iloc[:,0:2],data.iloc[:,2]
#iloc选取值，fillna()填充缺失值
inputs = inputs.fillna(inputs.mean(numeric_only=True))  #用均值填充缺失值
print(inputs)
```

对于类别值或离散值，可以将‘NaN’看作一个类别

```python
inputs= pd.get_dummies(inputs,dummy_na=True)  
# 将分类变量转化为one-hot编码(仅0/1)
# dummy_na=True表示将缺失值单独编码，即保留缺失值信息
print(inputs)
```
