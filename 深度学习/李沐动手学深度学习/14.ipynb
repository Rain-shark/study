# 数值稳定性 + 模型初始化和激活函数

## 数值稳定化

h 是层输出，t 表示层，y 是优化函数（包括损失函数）

损失关于参数的梯度由链式求导，所有的h都是向量，向量关于向量的导数是一个矩阵，所以有**很多矩阵相乘**

导致两个问题：

1. 梯度爆炸
2. 梯度消失

### 梯度爆炸



diag是对角矩阵  

对角矩阵的性质：矩阵 A 左乘一个对角矩阵 D，是分别用 D 的对角线元素分别作用于矩阵 A 的每一行，就相当于直接对矩阵A每一行数乘，所以可以用来表示激活函数的导数  



ReLU激活函数的导数为0或1

按上式可看出，结果为： 0/1 * 部分参数矩阵

参数矩阵层数越多，结果越大

梯度爆炸问题：

1. 梯度值超出值域
2. 对学习率敏感（梯度变化幅度）

### 梯度消失

最后结果是很多小数值的乘积，结果非常小

梯度消失的问题：

1. 梯度值为0
2. 训练无进展
3. 底层训练不好，仅顶层好

## 模型初始化和激活函数

让梯度在合理范围内：

1. 乘法变加法：ResNet\LSTM
2. 归一化：梯度归一化，梯度裁剪
3. 合理权重初始和激活函数

### 合理权重初始

随机初始化参数在合理值区间内对于小网络可以，但对于深度神经网络不合适

i 行 j 列 t 层，E 是均值，W 和 h 独立，r 是方差

要求输入和输出方差不变，所以有两个条件需要满足，难以同时满足，就取折中

### 激活函数

对于线性激活函数，如果要求均值为0，方差为1，则激活函数为 y = x

对于非线性激活函数：tanh 、 relu 、4*sigmoid - 2，都满足要求
